### Visualization with Seaborn

<div style="direction:rtl">
ایمپورت کتابخانه ها.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

<div style="direction:rtl">
ست کردن استایل.

```python
sns.set_style('whitegrid')
```

<div style="direction:rtl">   
لود کردن دیتا.

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

<div style="direction:rtl">
نمودار کرایه بر حسب سن را به شکل زیر رسم نمایید:
<div style="direction:rtl">
راهنمایی: نام ستون ترسیمی در کنار محور ها درج شده است.

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

<div style="direction:rtl">
نمودار توزیع آماری کرایه ها را به شکل رسم نمایید:

```python
sns.displot(titanic['fare'],aspect=1.5,height=4,color='red')
```

```python
plt.hist(titanic['fare'],color='red',bins=30)
```

<div style="direction:rtl">
نمودار جعبه ای کلاس بلیت را بر حسب سن رسم نمایید.

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
titanic[titanic['class']=='Third']['age'].max()
```

<div style="direction:rtl">
همان نمودار کلاس بر حسب سن  را با نمود تجمع افراد در سن های مختلف رسم نمایید.

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

<div style="direction:rtl">
نمودار تعداد افراد حاضر در کشتی را بر حسب جنسیت رسم نمایید.

```python
sns.catplot(x='sex',data=titanic,kind='count')
```

```python
sns.countplot(x='sex',data=titanic)
```

<div style="direction:rtl">
نمودار میزان همبستگی را در دیتاست تایتانیک رسم کنید و برای آن عنوان Titanic Correlations قرار دهید.

```python
sns.heatmap(titanic.corr(),cmap='Greens').set_title('Titanic Correlations')
sns.set_context('poster',font_scale=0.5)
```

<div style="direction:rtl">
نمودار توزیع آماری سن بر حسب جنسیت را در دو شبکه ی جدا با کمک سیبورن رسم نمایید.

```python
sns.displot(x='age',data=titanic,col='sex',height=3)
```

```python
sns.FacetGrid(data=titanic,col='sex').map(plt.hist,'age')
```

### Visualization with Pandas

<div style="direction:rtl">
در این بخش به سراغ ترسیم نمودار با خود پانداز می رویم.

```python
import pandas as pd
```

<div style="direction:rtl">   
لود کردن دیتا، دیتای مورد نظر برای شما ارسال خواهد شد.

```python
df3 = pd.read_csv('df3.txt')
```

<div style="direction:rtl">   
نگاه کلی به دیتاست، خلاصه ی دیتاست و خلاصه ی آماری دیتاست.

```python
df3.head()
```

```python
df3.info()
```

```python
df3.describe()
```

<div style="direction:rtl">   
نمودار نقطه ای دو ستون a و b را ترسیم نماید به شکلی که سایز نقاط آن50، رنگ آن قرمز و سایز قاب آن 12 در 3 باشد.

```python
df3.plot.scatter(x='a',y='b',figsize=(12,3),c='red',s=50)
```

<div style="direction:rtl">   
نمودار توزیع آماری ستون a را رسم نمایید.

```python
df3['a'].plot.hist()
```

<div style="direction:rtl">   
نمودار توزیع آماری ستون a را با استایل ggplot و تعداد ستون 25 و شفافیت(alpha) 0.5 رسم نمایید.

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=25,alpha=0.5)
df3['a'].plot.kde(alpha=1)
```

```python
sns.set_style('ggplot')
df3['a'].plot.hist(bins=25,alpha=0.5)
```

<div style="direction:rtl">   
نمودار جعبه ای ستون های a و b را رسم نمایید.
<div style="direction:rtl">   
نکته ی مهم : دقت فرمایید که در پانداز، نموداری جعبه ای را می توان برای خود ستون ترسیم نمود در حالی که در سیبورن یک ستون کتگوریکال لازم بود.

```python
df3[['a','b']].plot.box()
```

<div style="direction:rtl">   
تابع چگالی احتمال را برای ستون d ترسیم کنید.

```python
df3['d'].plot.density()
```

<div style="direction:rtl">   
همان تابع بالا را به شکل مقطع و با عرض 5 رسم نمایید.

```python
df3['d'].plot.density(lw=5,linestyle='--')
```

<div style="direction:rtl">   
نمودار دید کلی نسبت به ستون ها را برای 30 سطر اول ترسیم کنید. با شفافیت 0.4

```python
df3.head(30).plot.area(alpha=0.4)
```

<div style="direction:rtl">   
سعی کنید با سرچ و تحقیق، لجند پلات بالا را به شکل زیر در بیرون از نمودار قرار دهید.
<div style="direction:rtl">    
لینک زیر برای راهنمایی بیشتر قرار داده شده است:

[Legend Guide](https://matplotlib.org/3.3.2/tutorials/intermediate/legend_guide.html)

```python
df3.head(30).plot.area(alpha=0.4)
plt.legend(bbox_to_anchor=(1.05, 1))
```
