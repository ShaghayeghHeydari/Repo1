# Training and Testing Data

<div style="direction:rtl">
          (X)   (y)  .

```python
X=customers[['Length of Membership','Time on App','Time on Website','Avg. Session Length']]
Y=customers['Yearly Amount Spent']
```

<div style="direction:rtl">
     train_test_split     sklearn.model_selection            (70 )     (30 )    101(   )  . 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.3, random_state=101)
```

# Training the Model

<div style="direction:rtl">
     .
<div style="direction:rtl">
 sklearn.linear_model  LinearRegression   .

```python
from sklearn.linear_model import LinearRegression
```

<div style="direction:rtl">
              lm .

```python
lm = LinearRegression()
```

<div style="direction:rtl">
         .

```python
lm.fit(X_train, Y_train)
```

# Predicting Test Data

<div style="direction:rtl">
                                .                                         .

<div style="direction:rtl">
           :

```python
predictions= lm.predict(X_test)
```

# Evaluating the Model

<div style="direction:rtl">
             .            x=y    45     .

```python
plt.scatter(x= Y_test, y= predictions)
```

<div style="direction:rtl">
  Mean Absolute Error Mean Squared Error  the Root Mean Squared Error         .
<div style="direction:rtl">
         :

```python
customers['Yearly Amount Spent'].mean()
```

<div style="direction:rtl">
    500    (MAE)   10      2   .

<div style="direction:rtl">
        metrics   sklearn  .

```python
from sklearn import metrics
```

```python
print('MAE: ', metrics.mean_absolute_error(Y_test,predictions))
print('MSE: ', metrics.mean_squared_error(Y_test,predictions))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(Y_test,predictions)))
```
